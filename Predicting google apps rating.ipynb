<b> Data Imputation and Manipulation</b>

<i>fill the null values with appropiate values using aggregate functions such as: mean, median or mode</i>

```python
#define a function impute median, categorial value=mode and for numerical=median
def inpute_median(series):
    return series.fillna(series.median())
```

```python
google_data.Rating = google_data['Rating'].transform(inpute_median)
```

```python
google_data.isnull().sum()
```

```python
#mode of categorial values
print(google_data['Type'].mode())
print(google_data['Current Ver'].mode())
print(google_data['Android Ver'].mode())
```

```python
#fill the missing categorial values with mode
google_data['Type'].fillna(str(google_data['Type'].mode().values[0]), inplace=True)
google_data['Current Ver'].fillna(str(google_data['Current Ver'].mode().values[0]), inplace=True)
google_data['Android Ver'].fillna(str(google_data['Android Ver'].mode().values[0]), inplace=True)
```

```python
google_data.isnull().sum()
```

```python
#lets conver price,review and rating to the numerical values
google_data['Price'] = google_data['Price'].apply(lambda x: str(x).replace('$','')
                       if '$' in str(x) else(str(x)))
google_data['Price'] = google_data['Price'].apply(lambda x: float(x))
google_data['Reviews'] = pd.to_numeric(google_data['Reviews'], errors='coerce')#coerce means ignore
```

```python
google_data['Installs'] = google_data['Installs'].apply(lambda x: str(x).replace('+','')
                       if '+' in str(x) else(str(x)))
google_data['Installs'] = google_data['Installs'].apply(lambda x: str(x).replace(',','')
                       if ',' in str(x) else(str(x)))
google_data['Installs'] = google_data['Installs'].apply(lambda x: float(x))
```

<b> Data Visualization</b>

```python
grp = google_data.groupby('Category')
x = grp['Rating'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(12,5))
plt.title('Category Wise Rating')
plt.xlabel('Categories-->')
plt.ylabel('Rating--->')
plt.plot(x, 'ro')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.title('Category Wise Pricing')
plt.xlabel('Categories-->')
plt.ylabel('Prices--->')
plt.plot(y, 'r--', color='b')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.title('Category Wise Reviews')
plt.xlabel('Categories-->',color='b')
plt.ylabel('Reviews--->',color='b')
plt.plot(z, 'g^', color='g')#bs
plt.xticks(rotation=90)
plt.show()
```
